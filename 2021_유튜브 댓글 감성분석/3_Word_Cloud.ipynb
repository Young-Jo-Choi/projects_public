```python
import numpy as np
import pandas as pd
import json
```

```python
txts = []
with open('YouTubeResult.txt','r',encoding='UTF8') as f:
    string = f.readlines()
    for li in string:
        try:
            texts = json.loads(li)['text']
            text_list = texts.split(' ')
            txts.extend(text_list)
        except:
            pass
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from collections import Counter
tag_counts = Counter(txts)
```

```python
stopword = ['\n','아','ㅇ','더','다','수','','앙','그','거','ㅎㅎ','ㅠ','ㅠㅠ','ㅅㅂ','ㄷㄷ']
for count in range(20):
    stopword.append('ㅋ'*count)
```

```python
# stopword 옵션이 잘 먹히지 않음
fontpath = 'c:/Windows/Fonts/malgun.ttf'
worldcloud = WordCloud(background_color='white',
                      max_words=100,
                      relative_scaling=0.3,
                       font_path = fontpath,
                       stopwords=stopword,
                       width=1000,height=600
                      )
```

```python
worldcloud.generate_from_frequencies(tag_counts)
plt.imshow(worldcloud)
plt.axis('off')
plt.savefig('wordcloud_2.png')
```

```python
def remove_values_from_list(the_list, val):
    return [value for value in the_list if value != val]
```

```python
removed_txts = txts
for word in stopword:
    removed_txts = remove_values_from_list(removed_txts, word)
```

```python
from collections import Counter
tag_counts = Counter(removed_txts)
```

```python
worldcloud.generate_from_frequencies(tag_counts)
plt.imshow(worldcloud)
plt.axis('off')
plt.savefig('wordcloud.png')
```
